```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.histplot(df_groupby['gdp_per_capita'], kde=True, bins=10)

plt.title('Distribution of GDP per Capita')
plt.xlabel('GDP per Capita')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.scatterplot(data=df_groupby, x='total_cases', y='gdp_per_capita')

plt.title('Scatter Plot of Total Cases vs GDP per Capita')
plt.xlabel('Total Cases')
plt.ylabel('GDP per Capita')

plt.show()
```

```python
sns.pairplot(df_max)

# Show the plot
plt.show()
```

```python
# Create a bar plot using catplot
sns.catplot(x='continent', y='total_cases', data=df, kind='bar')

# Set the title and labels
plt.title('Total Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Cases')

# Show the plot
plt.show()
```
